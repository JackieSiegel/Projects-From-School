```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load 
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#View Data
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Go through variable/file grabbing data view data
player_demo = purchase_data.loc[:, ["Gender","SN", "Age"]]
player_demo.head()

#Drop duplicates and count players
player_demo = player_demo.drop_duplicates()
player_count = player_demo.count()[0]
player_count
```

```python
#Turn file into panda df 
pd.DataFrame({"Total Players": [player_count]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Basic Calculations
average_price = purchase_data["Price"].mean()
total_value = purchase_data["Price"].sum()
purchase_count = purchase_data["Price"].count()
item_count = len(purchase_data["Item ID"].unique())

#Create DataFrame to hold results
summary_df = pd.DataFrame({"Number of Unique Items": item_count,
                               "Total Revenue": [total_value],
                               "Number of Purchases": [purchase_count],
                               "Average Purchase Price": [average_price]})

#cleanup
summary_df = summary_df.round(2)
summary_df ["Average Price"] = summary_df["Average Purchase Price"].map("${:,.2f}".format)
summary_df ["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)
summary_df = summary_df.loc[:, ["Number of Unique Items", "Average Purchase Price", "Number of Purchases", "Total Revenue"]]

summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Basic Calculations
gender_demo_totals=player_demo['Gender'].value_counts()
gender_demo_percents = gender_demo_totals / player_count * 100

# Data Cleanup
gender_demo = pd.DataFrame({"Total Count": gender_demo_totals,
                                    "Percentage of Players": gender_demo_percents})

gender_demo = gender_demo.round(2)

#Data Gender_demo
gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Basic Calculations
gender_counts = purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")
gender_av = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
gender_ptotal = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")

# Cleanup
gender_data = pd.DataFrame({"Average Purchase Price": gender_av,
                            "Purchase Count": gender_counts, 
                            "Total Purchase Value": gender_ptotal})
#Display
gender_data
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Bin Ages
age_bins = [0, 9.90, 14.90, 19.90, 24.9, 29.9, 34.90, 39.90, 9999999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
del player_demo['SN']

#Categorize players
player_demo["Age Ranges"] = pd.cut(player_demo["Age"], age_bins, labels=group_names)
player_demo.head()

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Breakdown for percent
age_demo_totals = player_demo["Age Ranges"].value_counts()
age_demo_percents = age_demo_totals / player_count * 100
age_demo = pd.DataFrame({"Total Count": age_demo_totals, 
                       "Percent of Players": age_demo_percents})
age_demo = age_demo.sort_index()
age_demo
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Basic Calculations
user_total = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase Amount")
user_av = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
user_count = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")


user_data = pd.DataFrame({ "Purchase Count": user_count, 
                          "Average Purchase Price": user_av,
                        "Total Purchase Amount": user_total,})

# Convert to Data Frame
user_data.sort_values("Total Purchase Amount", ascending=False).head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items=purchase_data [["Item ID", "Item Name", "Price"]]
item_info=items.groupby(["Item ID","Item Name"])
purchase_count_item = item_info["Price"].count()
purchase_value = (item_info["Price"].sum()) 
item_price = purchase_value/purchase_count_item
most_popular_items = pd.DataFrame({"Item Price": item_price,
                                    "Purchase Count": purchase_count_item, 
                                   "Total Purchase Value":purchase_value})
popular_formatted = most_popular_items.sort_values(["Purchase Count"], ascending=False).head()
popular_formatted.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profit_df = pd.DataFrame(purchase_data)
profit_grp = profit_df.groupby(['Item ID','Item Name'])
profit_df = pd.DataFrame({"Purchase Count":profit_grp["Price"].count(),"Item Price":profit_grp["Price"].mean(),"Total Purchase Value":profit_grp["Price"].sum()})
profit_df = profit_df.sort_values("Total Purchase Value", ascending=False)
profit_df["Item Price"] = profit_df["Item Price"].map("${:.2f}".format)
profit_df["Total Purchase Value"] = profit_df["Total Purchase Value"].map("${:,.2f}".format)

profit_df.head()
```
